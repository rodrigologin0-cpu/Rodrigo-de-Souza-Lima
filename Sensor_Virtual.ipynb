```python
# ============================================================
# COMPARATIVO RNN x LSTM x TCN
# Sensor Virtual para Pressão do Forno
# Arquivo: Virtual_Sensor.xlsx
# Ambiente: Google Colab
# ============================================================

# Se necessário, rode uma vez:
# !pip install pandas numpy scikit-learn tensorflow openpyxl matplotlib joblib

import os
import time
import joblib
import unicodedata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers


# ============================================================
# CONFIGURAÇÕES PRINCIPAIS
# ============================================================

ARQUIVO_TREINO = "Virtual_Sensor.xlsx"

# Se tiver outro arquivo para teste final, coloque o nome aqui.
# Exemplo: ARQUIVO_NOVOS_DADOS = "Virtual_Sensor_Teste.xlsx"
ARQUIVO_NOVOS_DADOS = None

PASTA_SAIDA = "resultado_comparativo_rnn_lstm_tcn"
os.makedirs(PASTA_SAIDA, exist_ok=True)

# Se existir coluna de tempo, informe o nome.
# Se não existir, deixe None.
COL_TEMPO = None

# Se quiser forçar nomes exatos das colunas, coloque aqui.
# Se deixar None, o código tenta detectar automaticamente.
COL_POS_VALVULA = None
COL_VAZAO_TEORICA = None
COL_PRESSAO = None

# Exemplos manuais:
# COL_POS_VALVULA = "posicao_valvula"
# COL_VAZAO_TEORICA = "vazao_teorica"
# COL_PRESSAO = "pressao_controlada"

# Modelos que serão treinados
MODELOS_ATIVOS = ["RNN", "LSTM", "TCN"]

# Janela temporal
WINDOW_SIZE = 30

# Horizonte de previsão
# 1  = próxima amostra
# 10 = 10 passos à frente
HORIZON = 1

# Divisão cronológica treino/validação
TRAIN_RATIO = 0.70

# Normalização: "minmax" ou "standard"
NORMALIZACAO = "minmax"

# Tratamento de outliers:
# "clip"   = limita valores extremos
# "remove" = remove linhas com outliers
# "none"   = não trata outliers
OUTLIER_STRATEGY = "clip"

# Treinamento
EPOCHS = 5
BATCH_SIZE = 64
LEARNING_RATE = 1e-3

# Reprodutibilidade
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)


# ============================================================
# UNIDADES REAIS DO PROCESSO
# ============================================================

PRESSAO_MIN_REAL = 0
PRESSAO_MAX_REAL = 1600  # mmH2O

VALVULA_MIN_REAL = 0
VALVULA_MAX_REAL = 100  # %

VAZAO_MIN_REAL = 0
VAZAO_MAX_REAL = 100  # %

# Banda aceitável para erro de previsão
# Exemplo: previsão dentro de +/- 30 mmH2O
BANDA_ERRO_ACEITAVEL = 30


# ============================================================
# CONFIGURAÇÕES DOS GRÁFICOS
# ============================================================

# Número máximo de amostras exibidas no gráfico
N_AMOSTRAS_GRAFICO = 2000

# Opções:
# "inicio"    -> primeiras N amostras
# "fim"       -> últimas N amostras
# "intervalo" -> amostras entre AMOSTRA_INICIAL e AMOSTRA_FINAL
# "espacado"  -> N pontos distribuídos ao longo de toda a validação
MODO_GRAFICO = "inicio"

AMOSTRA_INICIAL = 0
AMOSTRA_FINAL = 2000

# Limites visuais ajustáveis para a pressão
MOSTRAR_LIMITES_PRESSAO = True
LIMITE_PRESSAO_INFERIOR = 0
LIMITE_PRESSAO_SUPERIOR = 1600

# Ajustar eixo Y automaticamente pela faixa definida
AJUSTAR_EIXO_Y_PELOS_LIMITES = True


# ============================================================
# UPLOAD DO ARQUIVO NO COLAB
# ============================================================

def garantir_arquivo_no_colab(caminho_arquivo):
    if os.path.exists(caminho_arquivo):
        print(f"Arquivo encontrado: {caminho_arquivo}")
        return

    print(f"Arquivo '{caminho_arquivo}' não encontrado no ambiente.")
    print("Faça upload do arquivo agora.")

    try:
        from google.colab import files
        uploaded = files.upload()

        if caminho_arquivo not in uploaded:
            print("\nArquivos enviados:")
            for nome in uploaded.keys():
                print("-", nome)

            raise FileNotFoundError(
                f"O arquivo enviado não tem o nome esperado: {caminho_arquivo}\n"
                f"Renomeie o arquivo ou altere ARQUIVO_TREINO no código."
            )

    except ImportError:
        raise FileNotFoundError(
            f"Arquivo não encontrado: {caminho_arquivo}\n"
            f"Coloque o Excel na mesma pasta do notebook ou informe o caminho correto."
        )


# ============================================================
# FUNÇÕES DE LEITURA E TRATAMENTO
# ============================================================

def normalizar_texto(txt):
    txt = str(txt).strip().lower()
    txt = unicodedata.normalize("NFKD", txt)
    txt = "".join([c for c in txt if not unicodedata.combining(c)])
    txt = txt.replace("_", " ")
    txt = txt.replace("-", " ")
    txt = " ".join(txt.split())
    return txt


def carregar_excel(caminho):
    df = pd.read_excel(caminho)
    df.columns = df.columns.astype(str).str.strip()

    if COL_TEMPO is not None:
        if COL_TEMPO not in df.columns:
            raise ValueError(
                f"A coluna de tempo '{COL_TEMPO}' não foi encontrada.\n"
                f"Colunas disponíveis: {list(df.columns)}"
            )

        df[COL_TEMPO] = pd.to_datetime(df[COL_TEMPO], errors="coerce")
        df = df.dropna(subset=[COL_TEMPO])
        df = df.sort_values(COL_TEMPO).reset_index(drop=True)

    return df


def detectar_coluna(df, candidatos_exatos, palavras_chave, nome_logico):
    colunas = list(df.columns)
    colunas_norm = {col: normalizar_texto(col) for col in colunas}

    candidatos_norm = [normalizar_texto(c) for c in candidatos_exatos]

    for col, col_norm in colunas_norm.items():
        if col_norm in candidatos_norm:
            return col

    for col, col_norm in colunas_norm.items():
        if all(palavra in col_norm for palavra in palavras_chave):
            return col

    raise ValueError(
        f"Não consegui detectar automaticamente a coluna de {nome_logico}.\n"
        f"Colunas disponíveis no Excel: {colunas}\n\n"
        f"Ajuste manualmente no começo do código."
    )


def resolver_colunas(df):
    print("\nColunas encontradas no Excel:")
    for col in df.columns:
        print("-", col)

    if COL_POS_VALVULA is None:
        col_pos = detectar_coluna(
            df,
            candidatos_exatos=[
                "posicao_valvula",
                "posição válvula",
                "posicao valvula",
                "posição da válvula",
                "posicao da valvula",
                "abertura valvula",
                "abertura da valvula",
                "valvula",
                "válvula"
            ],
            palavras_chave=["valvula"],
            nome_logico="posição da válvula"
        )
    else:
        col_pos = COL_POS_VALVULA

    if COL_VAZAO_TEORICA is None:
        col_vazao = detectar_coluna(
            df,
            candidatos_exatos=[
                "vazao_teorica",
                "vazão teórica",
                "vazao teorica",
                "vazão do forno",
                "vazao do forno",
                "vazao",
                "vazão"
            ],
            palavras_chave=["vazao"],
            nome_logico="vazão teórica"
        )
    else:
        col_vazao = COL_VAZAO_TEORICA

    if COL_PRESSAO is None:
        col_pressao = detectar_coluna(
            df,
            candidatos_exatos=[
                "pressao_controlada",
                "pressão controlada",
                "pressao controlada",
                "pressão",
                "pressao",
                "pv pressao",
                "pv pressão"
            ],
            palavras_chave=["pressao"],
            nome_logico="pressão controlada"
        )
    else:
        col_pressao = COL_PRESSAO

    for col in [col_pos, col_vazao, col_pressao]:
        if col not in df.columns:
            raise ValueError(
                f"A coluna '{col}' não foi encontrada.\n"
                f"Colunas disponíveis: {list(df.columns)}"
            )

    print("\nColunas selecionadas para o modelo:")
    print(f"Posição da válvula : {col_pos}")
    print(f"Vazão teórica      : {col_vazao}")
    print(f"Pressão controlada : {col_pressao}")

    return col_pos, col_vazao, col_pressao


def converter_para_numero(serie):
    if serie.dtype == "object":
        serie = (
            serie.astype(str)
            .str.replace(",", ".", regex=False)
            .str.replace(" ", "", regex=False)
        )

    return pd.to_numeric(serie, errors="coerce")


def tratar_outliers_iqr(df, colunas, estrategia="clip"):
    df_tratado = df.copy()
    mascara_valida = np.ones(len(df_tratado), dtype=bool)

    for col in colunas:
        q1 = df_tratado[col].quantile(0.25)
        q3 = df_tratado[col].quantile(0.75)
        iqr = q3 - q1

        if iqr == 0 or pd.isna(iqr):
            continue

        limite_inf = q1 - 1.5 * iqr
        limite_sup = q3 + 1.5 * iqr

        if estrategia == "clip":
            df_tratado[col] = df_tratado[col].clip(
                lower=limite_inf,
                upper=limite_sup
            )

        elif estrategia == "remove":
            mascara_col = (
                (df_tratado[col] >= limite_inf) &
                (df_tratado[col] <= limite_sup)
            )
            mascara_valida = mascara_valida & mascara_col

        else:
            raise ValueError(
                "OUTLIER_STRATEGY deve ser 'clip', 'remove' ou 'none'."
            )

    if estrategia == "remove":
        df_tratado = df_tratado[mascara_valida].copy()

    return df_tratado


def preparar_dataframe(df, feature_cols, target_col):
    df = df.copy()
    df.columns = df.columns.astype(str).str.strip()

    # Remove duplicatas lógicas.
    # A pressão entra como entrada e também como alvo.
    colunas_modelo = list(dict.fromkeys(feature_cols + [target_col]))

    colunas_necessarias = colunas_modelo.copy()

    if COL_TEMPO is not None:
        colunas_necessarias = [COL_TEMPO] + colunas_necessarias

    colunas_faltando = [
        col for col in colunas_necessarias
        if col not in df.columns
    ]

    if len(colunas_faltando) > 0:
        raise ValueError(
            f"As seguintes colunas não foram encontradas: {colunas_faltando}\n"
            f"Colunas disponíveis: {list(df.columns)}"
        )

    df = df[colunas_necessarias].copy()

    for col in colunas_modelo:
        df[col] = converter_para_numero(df[col])

    df[colunas_modelo] = df[colunas_modelo].interpolate(
        method="linear",
        limit_direction="both"
    )

    df = df.dropna(subset=colunas_modelo).reset_index(drop=True)

    if OUTLIER_STRATEGY != "none":
        df = tratar_outliers_iqr(
            df=df,
            colunas=colunas_modelo,
            estrategia=OUTLIER_STRATEGY
        )

    df = df.reset_index(drop=True)

    return df


# ============================================================
# JANELAS DESLIZANTES
# ============================================================

def criar_janelas_deslizantes(df, feature_cols, target_col, window_size, horizon):
    dados_x = df[feature_cols].values
    dados_y = df[target_col].values

    X = []
    y = []
    indices_y = []

    limite = len(df) - window_size - horizon + 1

    if limite <= 0:
        raise ValueError(
            f"Dados insuficientes para criar janelas.\n"
            f"Total de amostras: {len(df)}\n"
            f"WINDOW_SIZE: {window_size}\n"
            f"HORIZON: {horizon}\n"
            f"Mínimo necessário: {window_size + horizon} amostras."
        )

    for i in range(limite):
        inicio = i
        fim = i + window_size
        indice_alvo = fim + horizon - 1

        X.append(dados_x[inicio:fim, :])
        y.append(dados_y[indice_alvo])
        indices_y.append(indice_alvo)

    X = np.array(X)
    y = np.array(y).reshape(-1, 1)
    indices_y = np.array(indices_y)

    return X, y, indices_y


def dividir_treino_validacao(X, y, indices_y, train_ratio):
    n = len(X)
    n_treino = int(n * train_ratio)

    if n_treino <= 0 or n_treino >= n:
        raise ValueError(
            f"Divisão inválida.\n"
            f"Total de janelas: {n}\n"
            f"TRAIN_RATIO: {train_ratio}"
        )

    X_train = X[:n_treino]
    y_train = y[:n_treino]
    idx_train = indices_y[:n_treino]

    X_val = X[n_treino:]
    y_val = y[n_treino:]
    idx_val = indices_y[n_treino:]

    return X_train, X_val, y_train, y_val, idx_train, idx_val


# ============================================================
# NORMALIZAÇÃO
# ============================================================

def criar_scaler(tipo):
    if tipo == "minmax":
        return MinMaxScaler()
    elif tipo == "standard":
        return StandardScaler()
    else:
        raise ValueError("NORMALIZACAO deve ser 'minmax' ou 'standard'.")


def normalizar_janelas(X_train, X_val, y_train, y_val):
    n_train, janela, n_features = X_train.shape
    n_val = X_val.shape[0]

    scaler_x = criar_scaler(NORMALIZACAO)
    scaler_y = criar_scaler(NORMALIZACAO)

    X_train_2d = X_train.reshape(-1, n_features)
    X_val_2d = X_val.reshape(-1, n_features)

    scaler_x.fit(X_train_2d)
    scaler_y.fit(y_train)

    X_train_norm = scaler_x.transform(X_train_2d).reshape(
        n_train,
        janela,
        n_features
    )

    X_val_norm = scaler_x.transform(X_val_2d).reshape(
        n_val,
        janela,
        n_features
    )

    y_train_norm = scaler_y.transform(y_train)
    y_val_norm = scaler_y.transform(y_val)

    return X_train_norm, X_val_norm, y_train_norm, y_val_norm, scaler_x, scaler_y


def normalizar_janelas_novas(X_novo, scaler_x):
    n, janela, n_features = X_novo.shape
    X_2d = X_novo.reshape(-1, n_features)

    X_norm = scaler_x.transform(X_2d).reshape(
        n,
        janela,
        n_features
    )

    return X_norm


# ============================================================
# MODELOS: RNN, LSTM E TCN
# ============================================================

def construir_modelo_rnn(input_shape):
    entrada = layers.Input(shape=input_shape)

    x = layers.SimpleRNN(
        units=64,
        activation="tanh",
        return_sequences=False
    )(entrada)

    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.10)(x)

    saida = layers.Dense(1, activation="linear")(x)

    modelo = models.Model(inputs=entrada, outputs=saida)

    modelo.compile(
        optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
        loss="mse",
        metrics=["mae"]
    )

    return modelo


def construir_modelo_lstm(input_shape):
    entrada = layers.Input(shape=input_shape)

    x = layers.LSTM(
        units=64,
        activation="tanh",
        return_sequences=False
    )(entrada)

    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.10)(x)

    saida = layers.Dense(1, activation="linear")(x)

    modelo = models.Model(inputs=entrada, outputs=saida)

    modelo.compile(
        optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
        loss="mse",
        metrics=["mae"]
    )

    return modelo


def bloco_tcn(x, filtros, kernel_size, dilation_rate, dropout):
    residual = x

    x = layers.Conv1D(
        filters=filtros,
        kernel_size=kernel_size,
        padding="causal",
        dilation_rate=dilation_rate
    )(x)
    x = layers.LayerNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dropout(dropout)(x)

    x = layers.Conv1D(
        filters=filtros,
        kernel_size=kernel_size,
        padding="causal",
        dilation_rate=dilation_rate
    )(x)
    x = layers.LayerNormalization()(x)

    if residual.shape[-1] != filtros:
        residual = layers.Conv1D(
            filters=filtros,
            kernel_size=1,
            padding="same"
        )(residual)

    x = layers.Add()([x, residual])
    x = layers.Activation("relu")(x)

    return x


def construir_modelo_tcn(input_shape):
    entrada = layers.Input(shape=input_shape)

    x = entrada

    # TCN usando camadas nativas do Keras:
    # Conv1D causal + dilatação + bloco residual
    for dilatacao in [1, 2, 4, 8, 16]:
        x = bloco_tcn(
            x,
            filtros=64,
            kernel_size=3,
            dilation_rate=dilatacao,
            dropout=0.10
        )

    x = layers.GlobalAveragePooling1D()(x)

    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.10)(x)

    saida = layers.Dense(1, activation="linear")(x)

    modelo = models.Model(inputs=entrada, outputs=saida)

    modelo.compile(
        optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
        loss="mse",
        metrics=["mae"]
    )

    return modelo


def construir_modelo_por_nome(nome_modelo, input_shape):
    nome_modelo = nome_modelo.upper()

    if nome_modelo == "RNN":
        return construir_modelo_rnn(input_shape)

    elif nome_modelo == "LSTM":
        return construir_modelo_lstm(input_shape)

    elif nome_modelo == "TCN":
        return construir_modelo_tcn(input_shape)

    else:
        raise ValueError("Modelo inválido. Use: RNN, LSTM ou TCN.")


# ============================================================
# MÉTRICAS
# ============================================================

def calcular_smape(y_real, y_pred):
    y_real = y_real.flatten()
    y_pred = y_pred.flatten()

    denominador = (np.abs(y_real) + np.abs(y_pred)) / 2
    mascara = denominador > 1e-9

    if np.sum(mascara) == 0:
        return np.nan

    smape = np.mean(
        np.abs(y_real[mascara] - y_pred[mascara]) / denominador[mascara]
    ) * 100

    return smape


def avaliar_modelo(modelo, X, y_real_norm, scaler_y, nome_conjunto, nome_modelo):
    y_pred_norm = modelo.predict(X, verbose=0)

    # Volta para unidade original: mmH2O
    y_real = scaler_y.inverse_transform(y_real_norm)
    y_pred = scaler_y.inverse_transform(y_pred_norm)

    erro = y_real.flatten() - y_pred.flatten()
    erro_abs = np.abs(erro)

    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_real, y_pred)
    r2 = r2_score(y_real, y_pred)

    erro_medio = np.mean(erro)
    erro_std = np.std(erro)
    erro_max = np.max(erro_abs)
    erro_p95 = np.percentile(erro_abs, 95)
    erro_p99 = np.percentile(erro_abs, 99)

    nrmse_faixa = rmse / (PRESSAO_MAX_REAL - PRESSAO_MIN_REAL) * 100
    nmae_faixa = mae / (PRESSAO_MAX_REAL - PRESSAO_MIN_REAL) * 100

    dentro_banda = np.mean(erro_abs <= BANDA_ERRO_ACEITAVEL) * 100
    smape = calcular_smape(y_real, y_pred)

    if np.std(y_real.flatten()) > 0 and np.std(y_pred.flatten()) > 0:
        correlacao = np.corrcoef(y_real.flatten(), y_pred.flatten())[0, 1]
    else:
        correlacao = np.nan

    metricas = {
        "modelo": nome_modelo,
        "conjunto": nome_conjunto,
        "mse_mmH2O2": mse,
        "rmse_mmH2O": rmse,
        "mae_mmH2O": mae,
        "r2": r2,
        "correlacao": correlacao,
        "erro_medio_mmH2O": erro_medio,
        "desvio_erro_mmH2O": erro_std,
        "erro_max_mmH2O": erro_max,
        "erro_p95_mmH2O": erro_p95,
        "erro_p99_mmH2O": erro_p99,
        "nrmse_percentual_fundo_escala": nrmse_faixa,
        "nmae_percentual_fundo_escala": nmae_faixa,
        "smape_percentual": smape,
        f"percentual_erro_menor_{BANDA_ERRO_ACEITAVEL}_mmH2O": dentro_banda
    }

    print(f"\n===== {nome_modelo} | {nome_conjunto} =====")
    print(f"MSE                         : {mse:.3f} mmH2O²")
    print(f"RMSE                        : {rmse:.3f} mmH2O")
    print(f"MAE                         : {mae:.3f} mmH2O")
    print(f"R²                          : {r2:.5f}")
    print(f"Correlação                  : {correlacao:.5f}")
    print(f"Erro médio                  : {erro_medio:.3f} mmH2O")
    print(f"Erro P95                    : {erro_p95:.3f} mmH2O")
    print(f"Erro máximo                 : {erro_max:.3f} mmH2O")
    print(f"NRMSE fundo de escala       : {nrmse_faixa:.3f} %")
    print(f"Erro <= +/- {BANDA_ERRO_ACEITAVEL} mmH2O      : {dentro_banda:.2f} %")

    return y_real, y_pred, metricas


# ============================================================
# GRÁFICOS
# ============================================================

def selecionar_indices_grafico(
    y_real,
    n_amostras=N_AMOSTRAS_GRAFICO,
    modo=MODO_GRAFICO,
    amostra_inicial=AMOSTRA_INICIAL,
    amostra_final=AMOSTRA_FINAL
):
    total_amostras = len(y_real)

    if modo == "inicio":
        fim = min(n_amostras, total_amostras)
        indices = np.arange(0, fim)

    elif modo == "fim":
        inicio = max(0, total_amostras - n_amostras)
        indices = np.arange(inicio, total_amostras)

    elif modo == "intervalo":
        inicio = max(0, amostra_inicial)
        fim = min(amostra_final, total_amostras)
        indices = np.arange(inicio, fim)

    elif modo == "espacado":
        n_amostras = min(n_amostras, total_amostras)
        indices = np.linspace(
            0,
            total_amostras - 1,
            n_amostras
        ).astype(int)

    else:
        raise ValueError("modo deve ser: inicio, fim, intervalo ou espacado.")

    return indices


def plotar_real_vs_previsto(
    y_real,
    y_pred,
    titulo,
    nome_arquivo,
    nome_modelo,
    metricas=None,
    n_amostras=N_AMOSTRAS_GRAFICO,
    modo=MODO_GRAFICO,
    amostra_inicial=AMOSTRA_INICIAL,
    amostra_final=AMOSTRA_FINAL
):
    y_real = y_real.flatten()
    y_pred = y_pred.flatten()

    indices = selecionar_indices_grafico(
        y_real=y_real,
        n_amostras=n_amostras,
        modo=modo,
        amostra_inicial=amostra_inicial,
        amostra_final=amostra_final
    )

    x_plot = indices
    y_real_plot = y_real[indices]
    y_pred_plot = y_pred[indices]

    plt.figure(figsize=(16, 6))

    if MOSTRAR_LIMITES_PRESSAO:
        plt.axhspan(
            LIMITE_PRESSAO_INFERIOR,
            LIMITE_PRESSAO_SUPERIOR,
            alpha=0.08,
            label="Faixa operacional ajustável"
        )

        plt.axhline(
            LIMITE_PRESSAO_INFERIOR,
            linestyle="--",
            linewidth=1.2,
            label=f"Limite inferior: {LIMITE_PRESSAO_INFERIOR} mmH2O"
        )

        plt.axhline(
            LIMITE_PRESSAO_SUPERIOR,
            linestyle="--",
            linewidth=1.2,
            label=f"Limite superior: {LIMITE_PRESSAO_SUPERIOR} mmH2O"
        )

    plt.plot(
        x_plot,
        y_real_plot,
        label="Pressão real",
        linewidth=1.8
    )

    plt.plot(
        x_plot,
        y_pred_plot,
        label=f"Pressão prevista - {nome_modelo}",
        linewidth=1.5
    )

    if metricas is not None:
        texto_metricas = (
            f"RMSE: {metricas['rmse_mmH2O']:.2f} mmH2O\n"
            f"MAE: {metricas['mae_mmH2O']:.2f} mmH2O\n"
            f"R²: {metricas['r2']:.4f}\n"
            f"NRMSE: {metricas['nrmse_percentual_fundo_escala']:.2f}%\n"
            f"±{BANDA_ERRO_ACEITAVEL} mmH2O: "
            f"{metricas[f'percentual_erro_menor_{BANDA_ERRO_ACEITAVEL}_mmH2O']:.1f}%"
        )

        ax = plt.gca()
        ax.text(
            0.015,
            0.97,
            texto_metricas,
            transform=ax.transAxes,
            verticalalignment="top",
            bbox=dict(boxstyle="round", alpha=0.15)
        )

    plt.xlabel("Amostra da validação")
    plt.ylabel("Pressão controlada [mmH2O]")
    plt.title(titulo)
    plt.legend(loc="best")
    plt.grid(True, alpha=0.35)

    if AJUSTAR_EIXO_Y_PELOS_LIMITES:
        margem = 0.05 * (LIMITE_PRESSAO_SUPERIOR - LIMITE_PRESSAO_INFERIOR)
        plt.ylim(
            LIMITE_PRESSAO_INFERIOR - margem,
            LIMITE_PRESSAO_SUPERIOR + margem
        )

    plt.tight_layout()

    caminho = os.path.join(PASTA_SAIDA, nome_arquivo)
    plt.savefig(caminho, dpi=200)
    plt.show()

    print(f"Figura salva em: {caminho}")


def plotar_comparativo_modelos(resultados_validacao):
    plt.figure(figsize=(17, 7))

    primeiro_modelo = list(resultados_validacao.keys())[0]
    y_real = resultados_validacao[primeiro_modelo]["y_real"].flatten()

    indices = selecionar_indices_grafico(
        y_real=y_real,
        n_amostras=N_AMOSTRAS_GRAFICO,
        modo=MODO_GRAFICO,
        amostra_inicial=AMOSTRA_INICIAL,
        amostra_final=AMOSTRA_FINAL
    )

    x_plot = indices
    y_real_plot = y_real[indices]

    if MOSTRAR_LIMITES_PRESSAO:
        plt.axhspan(
            LIMITE_PRESSAO_INFERIOR,
            LIMITE_PRESSAO_SUPERIOR,
            alpha=0.08,
            label="Faixa operacional ajustável"
        )

        plt.axhline(
            LIMITE_PRESSAO_INFERIOR,
            linestyle="--",
            linewidth=1.1,
            label=f"Limite inferior: {LIMITE_PRESSAO_INFERIOR} mmH2O"
        )

        plt.axhline(
            LIMITE_PRESSAO_SUPERIOR,
            linestyle="--",
            linewidth=1.1,
            label=f"Limite superior: {LIMITE_PRESSAO_SUPERIOR} mmH2O"
        )

    plt.plot(
        x_plot,
        y_real_plot,
        label="Pressão real",
        linewidth=2.2
    )

    for nome_modelo, dados in resultados_validacao.items():
        y_pred = dados["y_pred"].flatten()
        y_pred_plot = y_pred[indices]

        plt.plot(
            x_plot,
            y_pred_plot,
            label=f"Previsão {nome_modelo}",
            linewidth=1.4
        )

    plt.xlabel("Amostra da validação")
    plt.ylabel("Pressão controlada [mmH2O]")
    plt.title(
        f"Comparativo de Previsão - RNN x LSTM x TCN | Horizonte = {HORIZON} passo(s)"
    )
    plt.legend(loc="best")
    plt.grid(True, alpha=0.35)

    if AJUSTAR_EIXO_Y_PELOS_LIMITES:
        margem = 0.05 * (LIMITE_PRESSAO_SUPERIOR - LIMITE_PRESSAO_INFERIOR)
        plt.ylim(
            LIMITE_PRESSAO_INFERIOR - margem,
            LIMITE_PRESSAO_SUPERIOR + margem
        )

    plt.tight_layout()

    caminho = os.path.join(PASTA_SAIDA, "comparativo_previsoes_rnn_lstm_tcn.png")
    plt.savefig(caminho, dpi=200)
    plt.show()

    print(f"Gráfico comparativo salvo em: {caminho}")


def plotar_convergencia_modelos(historicos):
    plt.figure(figsize=(14, 6))

    for nome_modelo, historico in historicos.items():
        plt.plot(
            historico.history["val_loss"],
            label=f"{nome_modelo} - validação"
        )

    plt.xlabel("Época")
    plt.ylabel("MSE normalizado")
    plt.title("Curva de Convergência - Validação")
    plt.legend(loc="best")
    plt.grid(True, alpha=0.35)
    plt.tight_layout()

    caminho = os.path.join(PASTA_SAIDA, "convergencia_validacao_modelos.png")
    plt.savefig(caminho, dpi=200)
    plt.show()

    print(f"Curva de convergência salva em: {caminho}")


def plotar_metricas_barras(df_metricas):
    df_val = df_metricas[df_metricas["conjunto"] == "Validação"].copy()
    df_val = df_val.sort_values("mse_mmH2O2")

    # MSE
    plt.figure(figsize=(10, 5))
    plt.bar(df_val["modelo"], df_val["mse_mmH2O2"])
    plt.xlabel("Modelo")
    plt.ylabel("MSE [mmH2O²]")
    plt.title("Comparativo de MSE - Validação")
    plt.grid(True, axis="y", alpha=0.35)
    plt.tight_layout()

    caminho = os.path.join(PASTA_SAIDA, "comparativo_mse_validacao.png")
    plt.savefig(caminho, dpi=200)
    plt.show()

    # RMSE
    plt.figure(figsize=(10, 5))
    plt.bar(df_val["modelo"], df_val["rmse_mmH2O"])
    plt.xlabel("Modelo")
    plt.ylabel("RMSE [mmH2O]")
    plt.title("Comparativo de RMSE - Validação")
    plt.grid(True, axis="y", alpha=0.35)
    plt.tight_layout()

    caminho = os.path.join(PASTA_SAIDA, "comparativo_rmse_validacao.png")
    plt.savefig(caminho, dpi=200)
    plt.show()


# ============================================================
# SALVAMENTO DE RESULTADOS
# ============================================================

def salvar_resultados_excel_modelo(y_real, y_pred, nome_modelo, nome_conjunto):
    df_resultado = pd.DataFrame({
        "pressao_real_mmH2O": y_real.flatten(),
        "pressao_prevista_mmH2O": y_pred.flatten(),
        "erro_mmH2O": y_real.flatten() - y_pred.flatten(),
        "erro_absoluto_mmH2O": np.abs(y_real.flatten() - y_pred.flatten())
    })

    nome_arquivo = f"resultado_{nome_conjunto.lower()}_{nome_modelo.lower()}.xlsx"
    caminho = os.path.join(PASTA_SAIDA, nome_arquivo)

    df_resultado.to_excel(caminho, index=False)

    return df_resultado


# ============================================================
# TREINAMENTO PRINCIPAL
# ============================================================

def executar_treinamento():
    garantir_arquivo_no_colab(ARQUIVO_TREINO)

    print("\nCarregando arquivo de treino...")
    df_bruto = carregar_excel(ARQUIVO_TREINO)

    col_pos, col_vazao, col_pressao = resolver_colunas(df_bruto)

    feature_cols = [
        col_pos,
        col_vazao,
        col_pressao
    ]

    target_col = col_pressao

    print("\nPreparando dados...")
    df = preparar_dataframe(
        df=df_bruto,
        feature_cols=feature_cols,
        target_col=target_col
    )

    print(f"\nTotal de amostras após tratamento: {len(df)}")

    print("\nPrimeiras linhas tratadas:")
    display(df.head())

    X, y, indices_y = criar_janelas_deslizantes(
        df=df,
        feature_cols=feature_cols,
        target_col=target_col,
        window_size=WINDOW_SIZE,
        horizon=HORIZON
    )

    print("\nFormato dos dados:")
    print(f"X: {X.shape}")
    print(f"y: {y.shape}")

    X_train, X_val, y_train, y_val, idx_train, idx_val = dividir_treino_validacao(
        X=X,
        y=y,
        indices_y=indices_y,
        train_ratio=TRAIN_RATIO
    )

    print("\nDivisão dos dados:")
    print(f"Janelas de treino    : {len(X_train)}")
    print(f"Janelas de validação : {len(X_val)}")

    X_train_norm, X_val_norm, y_train_norm, y_val_norm, scaler_x, scaler_y = normalizar_janelas(
        X_train=X_train,
        X_val=X_val,
        y_train=y_train,
        y_val=y_val
    )

    joblib.dump(scaler_x, os.path.join(PASTA_SAIDA, "scaler_x.pkl"))
    joblib.dump(scaler_y, os.path.join(PASTA_SAIDA, "scaler_y.pkl"))

    input_shape = X_train_norm.shape[1:]

    metricas_gerais = []
    historicos = {}
    modelos_treinados = {}
    resultados_validacao = {}

    for nome_modelo in MODELOS_ATIVOS:
        print("\n" + "=" * 80)
        print(f"TREINANDO MODELO: {nome_modelo}")
        print("=" * 80)

        tf.keras.backend.clear_session()

        modelo = construir_modelo_por_nome(
            nome_modelo=nome_modelo,
            input_shape=input_shape
        )

        modelo.summary()

        caminho_modelo = os.path.join(
            PASTA_SAIDA,
            f"melhor_modelo_{nome_modelo.lower()}.keras"
        )

        lista_callbacks = [
            callbacks.EarlyStopping(
                monitor="val_loss",
                patience=25,
                restore_best_weights=True
            ),
            callbacks.ReduceLROnPlateau(
                monitor="val_loss",
                factor=0.5,
                patience=10,
                min_lr=1e-6
            ),
            callbacks.ModelCheckpoint(
                filepath=caminho_modelo,
                monitor="val_loss",
                save_best_only=True
            )
        ]

        inicio_treinamento = time.time()

        historico = modelo.fit(
            X_train_norm,
            y_train_norm,
            validation_data=(X_val_norm, y_val_norm),
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            callbacks=lista_callbacks,
            verbose=1
        )

        tempo_treinamento = time.time() - inicio_treinamento

        melhor_epoca = int(np.argmin(historico.history["val_loss"]) + 1)
        menor_val_loss = float(np.min(historico.history["val_loss"]))

        modelo.save(
            os.path.join(PASTA_SAIDA, f"modelo_final_{nome_modelo.lower()}.keras")
        )

        historicos[nome_modelo] = historico
        modelos_treinados[nome_modelo] = modelo

        # Avaliação treino
        y_train_real, y_train_pred, metricas_train = avaliar_modelo(
            modelo=modelo,
            X=X_train_norm,
            y_real_norm=y_train_norm,
            scaler_y=scaler_y,
            nome_conjunto="Treino",
            nome_modelo=nome_modelo
        )

        metricas_train["tempo_treinamento_s"] = tempo_treinamento
        metricas_train["melhor_epoca"] = melhor_epoca
        metricas_train["menor_val_loss_normalizado"] = menor_val_loss

        # Avaliação validação
        y_val_real, y_val_pred, metricas_val = avaliar_modelo(
            modelo=modelo,
            X=X_val_norm,
            y_real_norm=y_val_norm,
            scaler_y=scaler_y,
            nome_conjunto="Validação",
            nome_modelo=nome_modelo
        )

        metricas_val["tempo_treinamento_s"] = tempo_treinamento
        metricas_val["melhor_epoca"] = melhor_epoca
        metricas_val["menor_val_loss_normalizado"] = menor_val_loss

        metricas_gerais.append(metricas_train)
        metricas_gerais.append(metricas_val)

        resultados_validacao[nome_modelo] = {
            "y_real": y_val_real,
            "y_pred": y_val_pred,
            "metricas": metricas_val
        }

        salvar_resultados_excel_modelo(
            y_real=y_val_real,
            y_pred=y_val_pred,
            nome_modelo=nome_modelo,
            nome_conjunto="validacao"
        )

        plotar_real_vs_previsto(
            y_real=y_val_real,
            y_pred=y_val_pred,
            titulo=f"{nome_modelo} - Previsão da Pressão | Horizonte = {HORIZON} passo(s)",
            nome_arquivo=f"grafico_validacao_{nome_modelo.lower()}.png",
            nome_modelo=nome_modelo,
            metricas=metricas_val
        )

    # Tabela geral
    df_metricas = pd.DataFrame(metricas_gerais)

    df_metricas = df_metricas.sort_values(
        by=["conjunto", "mse_mmH2O2"],
        ascending=[True, True]
    )

    caminho_metricas = os.path.join(
        PASTA_SAIDA,
        "comparativo_metricas_rnn_lstm_tcn.xlsx"
    )

    df_metricas.to_excel(caminho_metricas, index=False)

    print("\n" + "=" * 80)
    print("TABELA COMPARATIVA DE MÉTRICAS - TREINO E VALIDAÇÃO")
    print("=" * 80)
    display(df_metricas)

    df_metricas_validacao = df_metricas[
        df_metricas["conjunto"] == "Validação"
    ].copy()

    df_metricas_validacao = df_metricas_validacao.sort_values("mse_mmH2O2")

    print("\n" + "=" * 80)
    print("RANKING DOS MODELOS NA VALIDAÇÃO")
    print("=" * 80)

    colunas_post = [
        "modelo",
        "mse_mmH2O2",
        "rmse_mmH2O",
        "mae_mmH2O",
        "r2",
        "correlacao",
        "erro_p95_mmH2O",
        "nrmse_percentual_fundo_escala",
        f"percentual_erro_menor_{BANDA_ERRO_ACEITAVEL}_mmH2O",
        "melhor_epoca",
        "tempo_treinamento_s"
    ]

    display(df_metricas_validacao[colunas_post].round(4))

    caminho_metricas_post = os.path.join(
        PASTA_SAIDA,
        "metricas_para_post_linkedin.xlsx"
    )

    df_metricas_validacao[colunas_post].to_excel(
        caminho_metricas_post,
        index=False
    )

    print(f"\nTabela completa salva em: {caminho_metricas}")
    print(f"Tabela resumida para LinkedIn salva em: {caminho_metricas_post}")

    # Gráficos finais
    plotar_comparativo_modelos(resultados_validacao)
    plotar_convergencia_modelos(historicos)
    plotar_metricas_barras(df_metricas)

    melhor_modelo = df_metricas_validacao.iloc[0]["modelo"]

    print("\n" + "=" * 80)
    print("RESUMO FINAL")
    print("=" * 80)
    print(f"Melhor modelo pela métrica MSE na validação: {melhor_modelo}")
    print(f"Resultados salvos na pasta: {PASTA_SAIDA}")

    return {
        "modelos": modelos_treinados,
        "historicos": historicos,
        "scaler_x": scaler_x,
        "scaler_y": scaler_y,
        "feature_cols": feature_cols,
        "target_col": target_col,
        "df_tratado": df,
        "df_metricas": df_metricas,
        "resultados_validacao": resultados_validacao
    }


# ============================================================
# TESTE OPCIONAL COM NOVOS DADOS
# ============================================================

def testar_novos_dados(caminho_novos_dados, resultado_treinamento):
    if caminho_novos_dados is None:
        print("\nNenhum arquivo de novos dados informado.")
        return None

    if not os.path.exists(caminho_novos_dados):
        print(f"\nArquivo de novos dados não encontrado: {caminho_novos_dados}")
        return None

    modelos_treinados = resultado_treinamento["modelos"]
    scaler_x = resultado_treinamento["scaler_x"]
    scaler_y = resultado_treinamento["scaler_y"]
    feature_cols = resultado_treinamento["feature_cols"]
    target_col = resultado_treinamento["target_col"]

    print("\nCarregando novos dados para confirmação...")
    df_novo_bruto = carregar_excel(caminho_novos_dados)

    df_novo = preparar_dataframe(
        df=df_novo_bruto,
        feature_cols=feature_cols,
        target_col=target_col
    )

    X_novo, y_novo, indices_novo = criar_janelas_deslizantes(
        df=df_novo,
        feature_cols=feature_cols,
        target_col=target_col,
        window_size=WINDOW_SIZE,
        horizon=HORIZON
    )

    X_novo_norm = normalizar_janelas_novas(X_novo, scaler_x)
    y_novo_norm = scaler_y.transform(y_novo)

    metricas_novos = []
    resultados_novos = {}

    for nome_modelo, modelo in modelos_treinados.items():
        y_novo_real, y_novo_pred, metricas_novo = avaliar_modelo(
            modelo=modelo,
            X=X_novo_norm,
            y_real_norm=y_novo_norm,
            scaler_y=scaler_y,
            nome_conjunto="Novos Dados",
            nome_modelo=nome_modelo
        )

        metricas_novos.append(metricas_novo)

        resultados_novos[nome_modelo] = {
            "y_real": y_novo_real,
            "y_pred": y_novo_pred,
            "metricas": metricas_novo
        }

        salvar_resultados_excel_modelo(
            y_real=y_novo_real,
            y_pred=y_novo_pred,
            nome_modelo=nome_modelo,
            nome_conjunto="novos_dados"
        )

        plotar_real_vs_previsto(
            y_real=y_novo_real,
            y_pred=y_novo_pred,
            titulo=f"{nome_modelo} - Novos Dados | Horizonte = {HORIZON} passo(s)",
            nome_arquivo=f"grafico_novos_dados_{nome_modelo.lower()}.png",
            nome_modelo=nome_modelo,
            metricas=metricas_novo
        )

    df_metricas_novos = pd.DataFrame(metricas_novos)
    df_metricas_novos = df_metricas_novos.sort_values("mse_mmH2O2")

    caminho = os.path.join(PASTA_SAIDA, "metricas_novos_dados.xlsx")
    df_metricas_novos.to_excel(caminho, index=False)

    print("\nMétricas dos novos dados:")
    display(df_metricas_novos)

    return {
        "df_metricas_novos": df_metricas_novos,
        "resultados_novos": resultados_novos
    }


# ============================================================
# EXECUÇÃO
# ============================================================

resultado = executar_treinamento()

resultado_novos_dados = testar_novos_dados(
    caminho_novos_dados=ARQUIVO_NOVOS_DADOS,
    resultado_treinamento=resultado
)
```
